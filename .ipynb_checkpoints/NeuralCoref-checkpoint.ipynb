```python
import spacy
nlp = spacy.load('en_coref_lg')
doc = nlp(u"For centuries, the southeastern shores of the Baltic Sea were inhabited by various Baltic tribes. In the 1230s, the Lithuanian lands were united by Mindaugas, the King of Lithuania, and the first unified Lithuanian state, the Kingdom of Lithuania, was created on 6 July 1253. During the 14th century, the Grand Duchy of Lithuania was the largest country in Europe; present-day Lithuania, Belarus, Ukraine, and parts of Poland and Russia were the territories of the Grand Duchy. With the Lublin Union of 1569, Lithuania and Poland formed a voluntary two-state union, the Polishâ€“Lithuanian Commonwealth. The Commonwealth lasted more than two centuries, until neighbouring countries systematically dismantled it from 1772 to 1795, with the Russian Empire annexing most of Lithuania's territory. As World War I neared its end, Lithuanias Act of Independence was signed on 16 February 1918, declaring the founding of the modern Republic of Lithuania. In the midst of the Second World War, Lithuania was first occupied by the Soviet Union and then by Nazi Germany. As World War II neared its end and the Germans retreated, the Soviet Union reoccupied Lithuania. On 11 March 1990, a year before the formal dissolution of the Soviet Union, Lithuania became the first Baltic state[12] to declare itself independent, resulting in the restoration of an independent State of Lithuania after 50 years of Soviet occupation. Lithuania is a member of the European Union, the Council of Europe, a full member of the eurozone, Schengen Agreement, OECD and NATO. It is also a member of the Nordic Investment Bank, and part of Nordic-Baltic cooperation of Northern European countries. The United Nations Human Development Index lists Lithuania as a 'very high human development' country.")
```

```python
corefed = doc._.coref_resolved # Replaces coreferences with the reference - "It's -> Lithuania's"
```

```python
doc._.coref_clusters[0]
```

```python
cluster_names = []
for clust in doc._.coref_clusters:
    cluster_names.append(str(clust[0]).lower())
```

```python
cluster_names
```

```python
text_raw = corefed

import gensim

text = gensim.utils.simple_preprocess(text_raw)
```

```python
import pandas as pd
import numpy as np
text_clust = pd.DataFrame({
    'word': text
})
```

```python
def get_is_clust(word):
    if word in cluster_names:
        return 1
    else: 
        return 0
```

```python
text_clust['is_cluster'] = text_clust['word'].map(get_is_clust)
```

```python
def get_nearest_cluster(word_pos, n_nearest, text_clust, cluster_names):
    
    ind = word_pos
    ind_up, ind_dn = ind, ind

    match_count = 0
    cluster_match_count = np.repeat(0, len(cluster_names))

    # Up direction
    while ind_up >= 0:
        if text_clust.loc[ind_up]['is_cluster']:
            cluster_num = cluster_names.index(text_clust.loc[ind_up]['word'])
            cluster_match_count[cluster_num] += 1
            match_count += 1
            if match_count >= n_nearest:
                break
            
        ind_up -= 1

    # Down direction.
    while ind_dn < text_clust.shape[0]:
        if text_clust.loc[ind_dn]['is_cluster']:
            cluster_num = cluster_names.index(text_clust.loc[ind_dn]['word'])
            cluster_match_count[cluster_num] += 1
            match_count += 1
            if match_count >= n_nearest:
                break
        
        ind_dn += 1
            
    return cluster_match_count.argmax()
```

```python
word_clusters = []
for word_pos in range(len(text_clust['word'])):
    word_clusters.append(get_nearest_cluster(word_pos, n_nearest=4, text_clust=text_clust, cluster_names=cluster_names))
```

```python
text_clust['nearest_clusters'] = word_clusters
text_clust
```
